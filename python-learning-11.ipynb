```python
def gen_demo():

    yield "frist statement"
    yield "second statement"
    yield "third statement"
```

```python
gen=gen_demo()
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
gen=gen_demo()

for i in gen:
    print (i)
```

```python
def square(num):
    for i in range(1,num+1):
        yield i**2
```

```python
gen = square(10)

print(next(gen))
print(next(gen))
print(next(gen))

for i in gen:
    print (i)
```

```python
#range function using generator
def mera_range(start,end):

     for i in mera_range(start,end):
         yield i
```
